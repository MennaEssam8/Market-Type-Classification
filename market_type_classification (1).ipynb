```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
train=pd.read_csv("D:/DataSets/market type classification/train.csv")
test=pd.read_csv("D:/DataSets/market type classification/test.csv")
```

```python
train.head()
```

```python
train.describe()
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
print(train['X3'].unique())
```

```python
#fill missing values
train['X2'].fillna(train['X2'].mean(), inplace=True)
train['X9'].fillna(train['X9'].mode()[0], inplace=True)

train['X3'] = train['X3'].replace(['low fat', 'LF'], 'Low Fat')
train['X3'] = train['X3'].replace('reg', 'Regular')

mean_value = train['X4'][train['X4'] != 0].mean()
train['X4'] = train['X4'].replace(0, mean_value)
```

```python
train.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['X1'] = le.fit_transform(train['X1'])
```

```python
train['X3'] = le.fit_transform(train['X3'])
train['X5'] = le.fit_transform(train['X5'])
train['X7'] = le.fit_transform(train['X7'])
train['X9'] = le.fit_transform(train['X9'])
train['X10'] = le.fit_transform(train['X10'])
```

```python
train.head()
```

```python
x = train.drop('Y', axis=1)
y = train['Y'] 
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model =LogisticRegression(max_iter=200, solver='lbfgs')
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train_scaled, y_train, cv=10)  # 5-fold cross-validation
print("Cross-validation scores:", scores)
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
